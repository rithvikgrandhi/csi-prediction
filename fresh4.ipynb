```python
import matplotlib.pyplot as plt

import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.regularizers import l2
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import scipy.io

# Load the .mat file
file_path = './EV_Rank_1_52_RBs_50_UEs_1000_snaps.mat'
data = scipy.io.loadmat(file_path)

# Extract the relevant data
EV_data = data['EV_re_im_split']
data = EV_data
del EV_data
print(data.shape)

data = data[6]
print(data.shape)  # Should print (1000, 832)

feature_index = 8
timesteps = 1000

feature_data = data[:, feature_index]

# Plot the data
plt.figure(figsize=(16, 6))
plt.plot(range(timesteps), feature_data, linestyle='-', color='b')
plt.title(f'First {timesteps} Timesteps for Feature {feature_index}')
plt.xlabel('Timestep')
plt.ylabel('Feature Value')
plt.grid(True)
plt.show()
```
